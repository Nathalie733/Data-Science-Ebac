#### Bagging é um método de agrupamento que visa reduzir a variância de dados dentro de um conjunto. Com isso há o treinamento de vários modelos dentro de categorias de dados e agragando os as previsões através da agregação. 
#### O Bagging funciona através de três passos: Bootstrapping, Treinamento Paralelo e Agregação. O Bootstrapping gera diferentes subconjuntos a partir do conjunto de dados de treinamento ao selecionar pontos de forma aleatória; O Treinamento Paralelo treina as amostras independente e simultaneamente com outra unidade usando aprendizes fracos ou de base; e, a Agregação é uma média ou a maioria das previsões são tomadas para calcular uma estimativa mais precisa.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
data = datasets.load_wine(as_frame = True)

X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 22)
```

```python
dtree = DecisionTreeClassifier(random_state = 22)
dtree.fit(X_train,y_train)
```

```python
DecisionTreeClassifier(random_state=22)
```

```python
y_pred = dtree.predict(X_test)

print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred = dtree.predict(X_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred = y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
estimator_range = [2,4,6,8,10,12,14,16]
```

```python
models = []
scores = []

for n_estimators in estimator_range:

   
    clf = BaggingClassifier(n_estimators = n_estimators, random_state = 22)

   
    clf.fit(X_train, y_train)

    
    models.append(clf)
    scores.append(accuracy_score(y_true = y_test, y_pred = clf.predict(X_test)))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(9,6))
plt.plot(estimator_range, scores)

plt.xlabel("n_estimators", fontsize = 18)
plt.ylabel("score", fontsize = 18)
plt.tick_params(labelsize = 16)
plt.show()
```
