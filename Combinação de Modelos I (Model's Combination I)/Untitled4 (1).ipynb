# Hiperparâmetros de Random Forest:
* max_depth:  max_depth determina o número máximo de divisões que cada árvore pode realizar.
* criterion: medida de qualidade da divisão. Os valores possíveis são "gini" para o índice Gini e "entropy" para o ganho de informação.
* min_sample_split: min_sample_split determina o número mínimo de observações da árvore de decisão em qualquer nó para dividir.
* max_leaf_nodes: max_leaf_nodes restringe o crescimento de cada árvore.
* min_samples_leaf: min_samples_leaf especifica o número mínimo de amostras que devem estar presentes no nó folha após a divisão de um nó.
* n_estimators: n_estimator é o hiperparâmetro que define o número de árvores a serem utilizadas no modelo. A árvore também pode ser entendida como as subdivisões.
* max_sample (bootstrap sample): max_sample determina quanto do conjunto de dados é fornecido para cada árvore individual.
* max_features: Max_features limita uma contagem para selecionar o máximo de recursos em cada árvore.
* bootstrap: indica se a amostra deve ser criada com substituição (True) ou sem substituição (False).
* oob_score: indica se devemos calcular o score out-of-bag (True) ou não (False).
* n_jobs: número de trabalhos paralelos para executar durante o ajuste. O valor -1 significa usar todos os processadores disponíveis.
* random_state: estado aleatório usado para inicialização do gerador de números aleatórios. Esses hiperparâmetros podem ser ajustados usando a função RandomForestClassifier() do scikit-learn em Python.
