````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sklearn

np.random.seed(2360873)
N = 100
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df = df.drop('Unnamed: 0', axis=1)
df.head()
````

````python
sns.pointplot(x="posse_de_imovel", y="renda", data=df, dodge=True, ci = 90)
````

````python
sns.pointplot(x="posse_de_veiculo", y="renda", data=df, dodge=True, ci = 90)
````

````python
sns.pointplot(x="posse_de_imovel", y="renda", hue = 'posse_de_veiculo',
                   data=df, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

# As variáveis posse_de_veiculo e posse_de_imovel são sim fatores que demonstram o pode aquisitivo do cliente. Quanto maior sua renda, maior a probabilidade desse cliente ter um automóvel ou um imóvel.
