### I. Modelo no valor da gorjeta

````python
import pandas as pd
import numpy as np
from seaborn import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf

%matplotlib inline
````

````python
tips = load_dataset("tips")
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
tips.head()
````

````python
y, X = patsy.dmatrices('tip ~ sex + smoker + size + total_bill + 1', tips[:5])
X
````

````python
y
````

````python
reg_tip = smf.ols('tip ~ sex + smoker + time + net_bill', data=tips).fit()
tips['res_tip'] = reg_tip.resid
reg_tip.summary()
````

````python
reg_tip = smf.ols('tip ~ sex + net_bill', data=tips).fit()
tips['res_tip'] = reg_tip.resid
reg_tip.summary()
````

````python
plt.close('all')
sns.scatterplot(x='sex', y='res_tip', data=tips)
````

````python
reg_tip = smf.ols('tip ~ smoker + net_bill', data=tips).fit()
tips['res_tip'] = reg_tip.resid
reg_tip.summary()
````

````python
plt.close('all')
sns.scatterplot(x='smoker', y='res_tip', data=tips)
````

````python
plt.close('all')
sns.scatterplot(x='net_bill', y='res_tip', data=tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
reg_tip = smf.ols('tip ~ smoker + np.log(net_bill)', data=tips).fit()
tips['res_tip'] = reg_tip.resid
reg_tip.summary()
````

````python
reg_tip = smf.ols('tip ~ smoker + np.power(net_bill, 2)', data=tips).fit()
tips['res_tip'] = reg_tip.resid
reg_tip.summary()
````

````python
plt.close('all')

plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 3)

sns.regplot(ax = axes[0, 0],x=tips['tip'], y=tips['res_tip'], color='red');
sns.scatterplot(ax = axes[0, 1],x=tips['smoker'], y=tips['res_tip'], color='blue');
sns.scatterplot(ax = axes[0, 2],x=tips['sex'], y=tips['res_tip'], color='purple');
sns.regplot(ax = axes[1, 0],x=tips['net_bill'], y=tips['res_tip'], color='blue');
sns.countplot(ax = axes[1, 1],x=tips['time'], data=tips['res_tip']);
sns.scatterplot(ax = axes[1, 2],x=tips['total_bill'], y=tips['res_tip'], color='black');

plt.subplots_adjust(wspace=0.5, hspace=0.5)
````

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, diner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
reg_1 = smf.ols('tip ~ sex + smoker + time + net_bill', data=tips).fit()
tips['res_1'] = reg_1.resid
reg_1.summary()
````

````python
reg_2 = smf.ols('tip ~ smoker + net_bill', data=tips).fit()
tips['res_2'] = reg_2.resid
reg_2.summary()
````

````python
plt.close('all')
sns.scatterplot(x='net_bill', y='res_2', data=tips[tips['tip_pct'] < .5])
plt.axhline(y=0, color='r', linestyle='--')
````

````python
reg_3 = smf.ols('tip ~ smoker + np.log(net_bill)', data=tips).fit()
tips['res_3'] = reg_3.resid
reg_3.summary()
````

````python
reg_4 = smf.ols('tip ~ smoker + np.power(net_bill, 2)', data=tips).fit()
tips['res_4'] = reg_4.resid
reg_4.summary()
````

````python
print(f"R² do Modelo 1: {reg_1.rsquared*100:.2f}%")
print(f"R² do Modelo 3: {reg_3.rsquared*100:.2f}%")
````

````python
plt.close('all')

plt.rc('figure', figsize=(15, 15))
fig, axes = plt.subplots(2, 2)

sns.scatterplot(ax = axes[0, 0],x=tips['res_1'], y=tips['res_tip']);
sns.scatterplot(ax = axes[0, 1],x=tips['res_2'], y=tips['res_tip']);
sns.scatterplot(ax = axes[1, 0],x=tips['res_3'], y=tips['res_tip']);
sns.scatterplot(ax = axes[1, 1],x=tips['res_4'], y=tips['res_tip']);

plt.subplots_adjust(wspace=0.5, hspace=0.5)
````
